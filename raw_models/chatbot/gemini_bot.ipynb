````python
import google.generativeai as genai

API_KEY = "AIzaSyBWBlKUB4aDp36LZg0eAeQXHcf7LhEee20"
genai.configure(api_key=API_KEY)

# Load Gemini Model
model = genai.GenerativeModel("gemini-2.0-flash")


# 1️⃣ Function to Predict Injury Risk (Point-wise Format)
def predict_injury_risk(player_data):
    """
    Predicts the injury risk based on player's health and training data.
    Returns a point-wise list of risks, causes, and affected body parts.
    """
    prompt = f"""
    Analyze the injury risk for an athlete based on the following data:
    - Age: {player_data['age']}
    - Matches Played: {player_data['matches_played']}
    - Fatigue Level: {player_data['fatigue_level']}
    - Past Injuries: {player_data['past_injuries']}
    - Recovery Time after Injury: {player_data['recovery_time']}
    - Training Intensity: {player_data['training_intensity']}

    Generate the output in a **point-wise format** with **four elements per point**:

    🔹 **Format:**
    ```
    [Sl No] [Predicted Risk: High/Medium/Low] [Cause of Risk] [What it can affect]
    ```

    🔹 **Example Output:**
    ```
    [1] [High] [Excessive training load leading to fatigue] [Leg muscles, endurance]
    [2] [Medium] [Previous hamstring injury increases risk] [Hamstring, sprint speed]
    [3] [Low] [Good recovery history reduces risk] [N/A]

    **Summary**: The athlete has a high risk of muscle fatigue and hamstring strain due to overtraining.
    ```
    """

    response = model.generate_content(prompt)
    return response.text


# 2️⃣ Function for Recovery Assistance (With User Input for Severity)
def recovery_assistance(injury_type, age, past_injuries, sport):
    """
    Provides personalized recovery assistance based on:
    - Injury type
    - Athlete's age
    - Past injuries
    - Sport played
    - **User-inputted severity level** (1 = Low, 2 = Medium, 3 = High)
    """

    print("\n⚡ **Injury Severity Assessment** ⚡")
    print("Please perform the following tests and enter a severity level (1 = Low, 2 = Medium, 3 = High).")

    mobility = int(input("1️⃣ **Mobility Test**: Try moving the injured area slowly. Pain Level (1/2/3): "))
    pressure = int(input("2️⃣ **Pressure Test**: Apply light pressure. Pain Level (1/2/3): "))
    weight_bearing = int(input("3️⃣ **Weight-Bearing Test**: Try standing/applying weight. Pain Level (1/2/3): "))

    # Determine final severity level
    if mobility == 3 and pressure == 3 and weight_bearing == 3:
        severity_level = 3  # High
    elif max(mobility, pressure, weight_bearing) == 3:
        severity_level = 2  # Medium
    else:
        severity_level = 1  # Low

    print(f"\n✅ **Final Injury Severity Level:** {severity_level} (1 = Low, 2 = Medium, 3 = High)\n")

    # Step 2: Generate a personalized recovery plan based on injury details and severity
    recovery_prompt = f"""
    Generate a **personalized recovery plan** for an athlete with the following details:

    - **Injury Type**: {injury_type}
    - **Age**: {age} years
    - **Sport**: {sport}
    - **Past Injuries**: {past_injuries}
    - **Injury Severity Level**: {severity_level} (1 = Low, 2 = Medium, 3 = High)

    Provide:
    1️⃣ **Rehabilitation Plan** (Exercises suited for severity level)
    2️⃣ **Estimated Recovery Time**
    3️⃣ **Diet & Supplements for faster healing**
    4️⃣ **Precautions to prevent re-injury**

    Format the response clearly with headings and bullet points.
    """

    recovery_response = model.generate_content(recovery_prompt)

    return recovery_response.text


# 3️⃣ Function for Nutrition & Diet Plan
def diet_plan_suggestions(player_profile):
    """
    Suggests a customized diet plan for an athlete based on role, weight, and training needs.
    :param player_profile: Dictionary with 'sport', 'position', 'weight', 'height', 'goal'.
    :return: AI-generated diet plan.
    """
    prompt = f"""
    Suggest a **daily meal plan** for an athlete with the following details:
    - **Sport**: {player_profile['sport']}
    - **Position**: {player_profile['position']}
    - **Age**: {player_profile['age']}
    - **Weight**: {player_profile['weight']} kg
    - **Height**: {player_profile['height']} cm
    - **Training Goal**: {player_profile['goal']} (e.g., muscle gain, endurance, fat loss)

    Provide a structured meal plan including:
    - **Breakfast**
    - **Lunch**
    - **Dinner**
    - **Snacks**
    - **Macronutrient Breakdown (Protein, Carbs, Fats)**

    Make sure the diet aligns with the athlete's **specific needs**.
    """
    response = model.generate_content(prompt)
    return response.text


player_data = {
    "age": 20,
    "matches_played": 5,
    "fatigue_level": "Moderate",
    "past_injuries": "Ankle sprain",
    "recovery_time": "2 weeks",
    "training_intensity": "High"
}

player_profile = {
    "sport": "Basketball",
    "position": "Shooting Guard",
    "age": 20,
    "weight": 56,
    "height": 161,
    "goal": "Improve Agility & Speed"
}

print("\n🔹 Injury Risk Prediction:")
print(predict_injury_risk(player_data))

print("\n🔹 Recovery Assistance for Ankle Sprain:")
print(recovery_assistance("Ankle Sprain", 22, "Minor Knee Injury (2021)", "Basketball"))

print("\n🔹 Diet Plan for Basketball Shooting Guard:")
print(diet_plan_suggestions(player_profile))
````
